# Popularity Based Recomendation

```python
df = pd.merge(left=book_df,right=rating_df,on='ISBN',how='inner')
```

```python
df = pd.merge(left=df,right=user_df,on='User-ID',how='inner')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns='Age',inplace=True)
```

```python
df['Book-Title'].duplicated().sum()
```

```python
num_rating_df= df.groupby('Book-Title').count()[['Book-Rating']].reset_index().rename(columns={'Book-Rating':'num_ratings'})
```

```python
avg_rating_df = df.groupby('Book-Title')[['Book-Rating']].mean().reset_index().rename(columns={'Book-Rating':'avg_ratings'})
```

```python
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df,on="Book-Title")
```

```python
popularity_df
```

```python
top_50_df = popularity_df.loc[popularity_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
top_50_df = top_50_df.merge(book_df,on='Book-Title')
```

```python
top_50_df = top_50_df.drop_duplicates('Book-Title')
```

```python
type(top_50_df.head(1))
```

```python
final_df = top_50_df[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
# now this is the popularity based recomending system
final_df.head(5)
```

# collabrating filter based

```python
df.groupby('User-ID')['Book-Rating'].count()>200
```

```python
x = df.groupby('User-ID').count()['Book-Rating']>200
```

```python
x
```

```python
# this masking will remove all false values from serires and our objectives is to get userId who voted above 200
x = x[x]
```

```python
x
```

```python
# the index will contain all the user id who rated above 200 books
iq_users = x.index
```

```python
iq_users
```

```python
filtered_df = df[df['User-ID'].isin(iq_users)]
```

```python
 y = filtered_df.groupby('Book-Title').count()['Book-Rating'] > 50
```

```python
top_50_books = y[y].index
```

```python
filtered_df = filtered_df[filtered_df['Book-Title'].isin(top_50_books)]
```

```python
filtered_df.shape
```

```python
filtered_df.duplicated().sum()
```

```python
pt = filtered_df.pivot_table(index='Book-Title',columns='User-ID',values="Book-Rating")
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
type(pt)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(pt)
```

```python
len(similarity[1])
```

```python
sorted(list(enumerate(similarity[0])),key=lambda x:x[1],reverse=True)
```

```python
np.where(pt.index=='1984')
```

```python
def recommend_books(book_name):
    similar_movies = list()
    index_of_book = np.where(pt.index == book_name)[0][0]
    book_similarities = sorted(list(enumerate(similarity[index_of_book])),key=lambda x:x[1],reverse=True)[1:6]
    for i in book_similarities:
       similar_movies.append(pt.index[i[0]])
    return similar_movies
    
```
